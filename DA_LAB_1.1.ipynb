```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
insurance_data = pd.read_csv("/content/insurance.csv")
# Display the first few rows of the dataset
insurance_data.head()
```

```python
# Vectors
ages = insurance_data['age'].values
expenses = insurance_data['expenses'].values

# Factors (Categorical data)
insurance_data['smoker'] = insurance_data['smoker'].astype('category')
insurance_data['region'] = insurance_data['region'].astype('category')

# Lists
age_list = insurance_data['age'].tolist()
expenses_list = insurance_data['expenses'].tolist()

# Matrices
import numpy as np
age_expenses_matrix = np.column_stack((ages, expenses))

# Data frame
insurance_df = pd.DataFrame({
    'age': ages,
    'expenses': expenses
})

# Display the structures
print(f"Vector: {ages[:5]}")
print(f"List: {age_list[:5]}")
print(f"Matrix:\n{age_expenses_matrix[:5]}")
print(f"Data Frame:\n{insurance_df.head()}")
```

```python
# Scatter plot between age and smoking
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='expenses', hue='smoker', data=insurance_data)
plt.title('Scatter Plot between Age and Expenses based on Smoking Status')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```

```python
# Filter the data based on age (20 to 40)
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]

# Scatter plot between age (20 to 40) and smoking
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='expenses', hue='smoker', data=filtered_data)
plt.title('Scatter Plot between Age (20 to 40) and Expenses based on Smoking Status')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```

```python
# Box plot for expenses of smokers and non-smokers
plt.figure(figsize=(10, 6))
sns.boxplot(x='smoker', y='expenses', data=insurance_data)
plt.title('Expenses of Smokers vs. Non-Smokers')
plt.xlabel('Smoker')
plt.ylabel('Expenses')
plt.grid(True)
plt.show()
```

```python
# Count plot for number of smokers gender-wise
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='smoker', data=insurance_data)
plt.title('Number of Smokers Gender-wise')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```
